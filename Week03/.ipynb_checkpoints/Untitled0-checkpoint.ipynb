```python
import pandas as pd
import sqlite3
from pandas.io import sql
from scipy import stats
import statsmodels.formula.api as smf

conn = sqlite3.connect('enron.db')
```

```python
sql_txt = """select name, seniority, mid, max(rno) rcpt_count
from EmployeeBase join MessageBase 
on eid = from_eid 
join RecipientBase using (mid)
group by name, seniority, mid 
"""

messages = sql.frame_query(sql_txt, conn)
```

```python
print messages.head()
```

```python
messages.boxplot(column='rcpt_count', by='seniority')
```

```python
seniority_means = messages.groupby(['seniority']).mean()
print seniority_means
```

```python
messages.groupby(['seniority', 'rcpt_count']).count()
```

```python
messages.groupby('seniority').hist(column=['rcpt_count'], alpha=0.5)
```

```python
test = messages[messages['seniority'] == 'Junior']
test.head()
```

```python
# this is how the internet told me how to do it
z_stat, p_val = stats.ranksums(messages[messages['seniority'] == 'Junior']['rcpt_count'], messages[messages['seniority'] == 'Senior']['rcpt_count'])

print p_val
# at 5%, can't really conclude that rcpt count differs

```

```python
lm = smf.ols('rcpt_count ~ seniority', data=messages)
fit = lm.fit()
print fit.summary()
```
