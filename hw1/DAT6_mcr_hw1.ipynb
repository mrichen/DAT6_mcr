```python
%matplotlib inline

import pandas as pd
import sqlite3
from pandas.io import sql
conn = sqlite3.connect('enron.db')
```

```python
msg_length_df = sql.frame_query("select mid, datetime(unix_time, 'unixepoch') as msg_dt, length(subject) as sublen from MessageBase", conn, "mid")
msg_length_df.head()
```

```python
msg_length_df.plot() # test w/ all data points first
```

```python
wb = msg_length_df.msg_dt.map(lambda x: pd.Period(x, freq='W'))
msg_length_df.groupby(wb).sublen.median().plot() # plot bucketed output
```
